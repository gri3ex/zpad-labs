# Завдання 1: Завантаження датасету та Data Cleaning

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
columns = ["symboling", "normalized-losses", "make", "fuel-type", "aspiration", "num-of-doors", "body-style", 
           "drive-wheels", "engine-location", "wheel-base", "length", "width", "height", "curb-weight", 
           "engine-type", "num-of-cylinders", "engine-size", "fuel-system", "bore", "stroke", "compression-ratio", 
           "horsepower", "peak-rpm", "city-mpg", "highway-mpg", "price"]

df = pd.read_csv(url, names=columns)

# Очищення
df.replace("?", pd.NA, inplace=True)
df.dropna(subset=['price', 'horsepower'], inplace=True)
df['price'] = pd.to_numeric(df['price'])
df['horsepower'] = pd.to_numeric(df['horsepower'])

df.head()
```

Scatter Plot

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='horsepower', y='price')
plt.title('Залежність ціни від потужності')
plt.show()
```

Histogram

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=15, kde=True, color='blue')
plt.title('Розподіл цін')
plt.show()
```

Box Plot або Bar Chart

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='body-style', y='price')
plt.title('Ціна за типом кузова')
plt.show()
```

Heatmap

```python
plt.figure(figsize=(12, 8))
# тільки числові дані для кореляції
numeric_df = df.select_dtypes(include=['float64', 'int64'])
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title('Кореляція характеристик')
plt.show()
```
